```python
import os
import cv2
```

```python

# 1 horizontally filpped the image - water image
def horizontal_flip(image_folder_path, output_folder, image_name):

    i = 1
    for image_path in os.listdir(image_folder_path):

        image = cv2.imread(f"{image_folder_path}/{image_path}")
        image = cv2.resize(image, (640, 640))
        
        h_flip = cv2.flip(image, 1)

        cv2.imwrite(f"{output_folder}/{image_name}-{i}-1.jpg", h_flip)

        print(f"{i} image write h flip done")
        i += 1

        
# 2 vetically flipped image - mirror image
def vertical_flip(image_folder_path, output_folder, image_name):
    i = 1
    for image_path in os.listdir(image_folder_path):
        image = cv2.imread(f"{image_folder_path}/{image_path}")

        image = cv2.resize(image, (640, 640))

        v_flip = cv2.flip(image, 0)

        cv2.imwrite(f"{output_folder}/{image_name}-{i}-2.jpg", v_flip)

        print(f"{i} image write v flip done")
        i += 1

# 3 rotation 30 degree
def rotate_30(image_folder_path, output_folder, image_name):
    i = 1
    for image_path in os.listdir(image_folder_path):
        image = cv2.imread(f"{image_folder_path}/{image_path}")
        image = cv2.resize(image, (640, 640))

        (h, w) = image.shape[:2]
        center = (w // 2, h // 2)
        M = cv2.getRotationMatrix2D(center, 30, 1.0)  # 30-degree rotation
        # rotated_image_30 = cv2.warpAffine(image, M, (w, h))
        rotated_image_30 = cv2.warpAffine(image, M, (w, h))

        cv2.imwrite(f"{output_folder}/{image_name}-{i}-3.jpg", rotated_image_30)

        print(f"{i} rotation 30 done")

        i += 1


# 4 rotation 25 degree
def rotate_25(image_folder_path, output_folder, image_name):
    i = 1
    for image_path in os.listdir(image_folder_path):
        image = cv2.imread(f"{image_folder_path}/{image_path}")
        # image = cv2.resize(image, (224, 224))
        image = cv2.resize(image, (640, 640))

        (h, w) = image.shape[:2]
        center = (w // 2, h // 2)
        M = cv2.getRotationMatrix2D(center, 25, 1.0)  # 25-degree rotation
        # rotated_image_25  = cv2.warpAffine(image, M, (w, h))
        rotated_image_25 = cv2.warpAffine(image, M, (w, h))

        cv2.imwrite(f"{output_folder}/{image_name}-{i}-4.jpg", rotated_image_25)

        i += 1
        
# 5 rotation 75 degree
def rotate_75(image_folder_path, output_folder, image_name):
    i = 1
    for image_path in os.listdir(image_folder_path):
        image = cv2.imread(f"{image_folder_path}/{image_path}")
        # image = cv2.resize(image, (224, 224))
        image = cv2.resize(image, (640, 640))

        (h, w) = image.shape[:2]
        center = (w // 2, h // 2)
        M = cv2.getRotationMatrix2D(center, 75, 1.0)  # 75-degree rotation
        # rotated_image_75 = cv2.warpAffine(image, M, (w, h))
        rotated_image_75 = cv2.warpAffine(image, M, (w, h))

        cv2.imwrite(f"{output_folder}/{image_name}-{i}-5.jpg", rotated_image_75)

        # print(f"{i} rotation 75 done")

        i += 1
```

```python
# image augmentation


# Tags on image first number means image number and second number represent the type of augmentation
# EX - data/lemon-balm-augment/lemon-balm-1-5 : means first image, rotated 75 degree augment image
# 1 - horizontal flip
# 2 - vertical flip
# 3 - ratated 30 degree
# 4 - rotated 25 degree
# 5 - rotated 75 degree

image_folder_path = "./data/spear_mint"
output_folder = "./data/_augmented_images/spear_mint_augment"
image_name = "spear_mint"

while True:
    augment_type= input("augment")

    if augment_type == '':
        break

    print(f"augment type - {augment_type}")

    if augment_type == '1':

        horizontal_flip(image_folder_path, output_folder, image_name)

    elif augment_type == '2':
        vertical_flip(image_folder_path, output_folder, image_name)

    elif augment_type == '3':
        print("you have selected 30 degree rotate")
        rotate_30(image_folder_path, output_folder, image_name)

        print("30 degree rotation all images done")
    elif augment_type == '4':
        print("25 deg rotate")
        rotate_25(image_folder_path, output_folder, image_name)
        print(" 25 deg rotate done")
    elif augment_type == '5':
        print("75 deg rotate")
        rotate_75(image_folder_path, output_folder, image_name)
        print(" 75 deg rotate done")


```
